```python
# средняя квадратичная ошибка
def mean_square_error(b_0, b_1, x, y):
    sum_error = sum(error(b_0, b_1, x_i, y_i) ** 2 
                for x_i, y_i in zip(x, y))
    mean_error = (1/len(x))*sum_error
    return mean_error
```

```python
import re
from collections import defaultdict
```

```python
# сгенерировать слова без повтор
def tokenize(message):
    message = message.lower()
    all_words = re.findall("[а-яa-z0-9']+", message)
    return set(all_words)
```

```python
# Посчитать частность слов
def count_words(training_set):
    counts = defaultdict(lambda: [0, 0])
    for label, message in training_set:
            for word in tokenize(message):
                counts[word][0 if label == 'spam' else 1] += 1
    return counts
```

```python
def word_probabilities(counts, total_spams, total_non_spams, k=0.5):
    return [(w, (spam + k) / (total_spams + 2 * k),
           (non_spam + k) / (total_non_spams + 2 * k))
           for w, (spam, non_spam) in counts.items()]
```

```python
import math

def spam_probability(words_probs, message):
    message_words = tokenize(message)
    log_prob_if_spam = log_prob_if_not_spam = 0.0
    
    for word, prob_if_spam, prob_if_not_spam in words_probs:
        
        if word in message_words:
            log_prob_if_spam += math.log(prob_if_spam)
            log_prob_if_not_spam += math.log(prob_if_not_spam)
            
        else:
            log_prob_if_spam += math.log(1.0 - prob_if_spam)
            log_prob_if_not_spam += math.log(1.0 - prob_if_not_spam)
            
        prob_if_spam = math.exp(log_prob_if_spam)
        prob_if_not_spam = math.exp(log_prob_if_not_spam)
        
    return prob_if_spam / (prob_if_spam + prob_if_not_spam)
```

```python
class NaiveBayesClassifier:
    
    def __init__(self, k=0.5):
        self.k = k
        self.words_probs = []
        
    # Обучить классификатор при помощи обучающей выборки
    def train(self, training_set):
        
        # посчитать спамные и неспамные сообщения
        total_spams = 0
        total_non_spams = 0

        for i in range(len(data)):
            if data[i][0] == 'ham':
                total_non_spams += 1
            else:
                total_spams += 1
        
        # пропустить обучающую выборку через контейнер
        words_counts = count_words(training_set)
        self.words_probs = word_probabilities(words_counts, total_spams, total_non_spams, self.k)
        
    def classify(self, message):
        return spam_probability(self.words_probs, message)
```
