```python
import matplotlib.pyplot as plt 
import pandas as pd 

data = pd.read_csv("/content/data_csv.txt", sep=',', names = ['current motion','sound level','illuminance','temperature'])
data.head(5)
```

```python
motion = list(data['current_motion'])
sound = list(data['sound level'])
illuminance = list(data['illuminance'])
temperature = list(data['temperature'])

fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)

ax.set_xlim([0,len(motion)])
ax.set_ylim([0, max(motion) + 20])
ax.set_title('Motion')
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.plot(motion)

plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)

ax.set_xlim([0,len(sound)])
ax.set_ylim([0, max(sound) + 5])
ax.set_title('Sound level')
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.plot(sound)

plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)

ax.set_xlim([0,len(illuminance)])
ax.set_ylim([250, max(illuminance) + 10])
ax.set_title('Illuminance')
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.plot(illuminance)

plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)

ax.set_xlim([0,len(temperature)])
ax.set_ylim([25.8, 26.2])
ax.set_title('Temperature')
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.plot(temperature)

plt.grid()
plt.show()
```

```python
import matplotlib.ticker as ticker
fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)

ax.set_title('Illuminance')
ax.set_xlabel('Areas of the values')
ax.set_ylabel('Count of records')

ax.hist(illuminance, [310, 320], color='b')
ax.hist(illuminance, [320, 330], color='b')
ax.hist(illuminance, [330, 340], color='b')
ax.hist(illuminance, [340, 350], color='b')
ax.hist(illuminance, [350, 360], color='b')
ax.hist(illuminance, [360, 370], color='b')
ax.hist(illuminance, [370, 380], color='b')

plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(25, 10))
ax = fig.add_subplot(111)
fig.patch.set_facecolor('white')


a = [0,0,0,0,0]

for i in range(len(illuminance)):
    if illuminance[i] < 330:
        a[0] += 1
    elif illuminance[i] >= 330 and illuminance[i] < 340:
        a[1] += 1
    elif illuminance[i] >= 340 and illuminance[i] < 350:
        a[2] += 1
    elif illuminance[i] >= 350 and illuminance[i] < 360:
        a[3] += 1
    elif illuminance[i] >= 360:
        a[4] += 1

ax.set_title('Illuminance')
plt.pie(a, labels=['<330','330-340','340-350','350-360','360>'], autopct='%1.2f%%')
plt.show()
```
