```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, accuracy_score

from sklearn.model_selection import GridSearchCV

from matplotlib import pyplot as plt
import numpy as np
```

```python
X,y = make_moons(n_samples=1000, noise=0.25)
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

Solving the task of classification

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
print('mean_squared_error: ' + str(mean_squared_error(y_test, knn.predict(X_test))))
print('accuracy_score: ' + str(accuracy_score(y_test, knn.predict(X_test))))
```

Plot a hyperfield

```python
x_space = np.linspace(-2, 2, 100)
x_grid, y_grid = np.meshgrid(x_space, x_space)
```

```python
xy = np.stack([x_grid, y_grid], axis=2).reshape(-1, 2) 
xy
```

```python
plt.scatter(xy[:, 0], xy[:, 1], s=1, alpha=0.3, c=knn.predict(xy))
plt.scatter(X_train[:,0], X_train[:,1], c = y_train)
```
